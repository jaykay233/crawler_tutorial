cookie的处理方式
- 手动处理  
    - 将cookie粘贴到headers中
- 自动处理
    - 基于session对象实现自动处理，可以通过requests.Session()得到一个session。该对象可以像requests一样进行get和post，只不过会生成cookie，下次如果再用该session请求的话，就还会使用同一个cookie
    - 至少会使用到2次session对象么，第一次是为了捕获cookie并且将其存储下来，第二次是为了携带cookie发送请求

```python
import lxml 
from lxml import etree
```

```python
url = 'https://www.89ip.cn/'
http_proxies=[]
for i in range(1,10):
    if i == 1:
        new_url = url
    else:
        new_url = url + 'index_%d.html'%(i)
    response = requests.get(url=new_url,headers=headers)
    page_text = response.text
    tree = etree.HTML(page_text)
    tr_list = tree.xpath('//tr' )[1:]
    for tr in tr_list:
        ip_port=tr[0].text.strip()+':'+tr[1].text.strip()
        http_proxies.append({'http':ip_port})
```
